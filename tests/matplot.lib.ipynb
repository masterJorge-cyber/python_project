```python
import pandas as pd
np.random.seed(19680801)

data = pd.DataFrame(
    {'a': np.arange(50), #0 a 50
     'c': np.random.randint(0, 50, 50), #Gera 50 nmeros inteiros aleatrios entre 0 (inclusivo) e 50 (exclusivo).
     'd': np.random.randn(50)}) #Gera 50 nmeros aleatrios da distribuio normal padro (mdia = 0, desvio padro = 1).

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
print(data)

fig, ax =  plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```
